OBJECTIVE:1 -- Monitor Outbreak Trends and Hotspots
Track confirmed cases, suspected cases, deaths, and weekly new cases by country.
Identify countries with the highest case fatality rates (CFRs).
Spotlight regions where clade distribution may influence severity 

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Aggregate like SQL
country_summary = (
    mpox.groupby("Country")
    .agg(
        total_suspected=("Suspected_Cases", "sum"),
        total_confirmed=("Confirmed_Cases", "sum"),
        avg_weekly_cases=("Weekly_New_Cases", "mean"),
        total_deaths=("Deaths", "sum")
    )
    .reset_index()
)
```

```python
# CFR calculation
country_summary["cfr_percent"] = (country_summary["total_deaths"] /
                                  country_summary["total_confirmed"].replace(0, pd.NA)) * 100
```

```python
# Sort top 5
top_countries = country_summary.sort_values("total_confirmed", ascending=False).head(5)

# --- Visualization ---
plt.figure(figsize=(10,6))
ax = sns.barplot(
    x="Country", 
    y="cfr_percent", 
    data=top_countries, 
    hue="Country",          # set hue
    palette="Reds_r", 
    legend=False            # hide extra legend
)

# Add CFR% values on top of each bar
for container in ax.containers:
    ax.bar_label(container, fmt="%.2f%%", label_type="edge", padding=3)

plt.title("Case Fatality Rate (CFR) for Top 5 Countries by Confirmed Cases")
plt.ylabel("CFR (%)")
plt.xlabel("Country")
plt.tight_layout()
plt.show()
```

```python
clade_summary = (
    mpox.groupby(["Country", "Clade"])
    .agg(
        total_confirmed=("Confirmed_Cases", "sum"),
        total_deaths=("Deaths", "sum")
    )
    .reset_index()
)

clade_summary["cfr_percent"] = (clade_summary["total_deaths"] /
                                clade_summary["total_confirmed"].replace(0, pd.NA)) * 100

# Top clade-country pairs by CFR
top_clade_cfr = clade_summary.sort_values("cfr_percent", ascending=False).head(5)

# --- Visualization ---
plt.figure(figsize=(12,6))
sns.barplot(x="Clade", y="cfr_percent", hue="Country", data=top_clade_cfr)
plt.title("Top Clade-Country Combinations by CFR")
plt.ylabel("CFR (%)")
plt.show()
```

```python
unknown_clades = (
    mpox[mpox["Clade"].isna() | mpox["Clade"].str.contains("Unknown", case=False, na=False)]
    .groupby("Country")
    .agg(
        total_confirmed=("Confirmed_Cases", "sum"),
        total_deaths=("Deaths", "sum")
    )
    .reset_index()
)

unknown_clades["cfr_percent"] = (unknown_clades["total_deaths"] /
                                 unknown_clades["total_confirmed"].replace(0, pd.NA)) * 100

# --- Visualization ---
plt.figure(figsize=(10,6))
sns.barplot(x="Country", y="total_confirmed",hue="Country", data=unknown_clades, palette="gray")
plt.title("Countries with Unknown Clade Reporting")
plt.ylabel("Confirmed Cases")
plt.show()
```

OBJECTIVE 2: --Assess Vaccination Progress and Gaps
Compare vaccine doses allocated vs deployed vs administered.
Evaluate vaccine coverage by country.
Highlight countries with high allocations but low deployment/uptake.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Aggregate vaccine data per country
vaccine_summary = (
    mpox.groupby("Country")
    .agg(
        allocated=("Vaccine_Dose_Allocated", "sum"),
        deployed=("Vaccine_Dose_Deployed", "sum"),
        administered=("Vaccinations_Administered", "sum"),
        latest_coverage=("Vaccine_Coverage", "max")  # take latest coverage value
    )
    .reset_index()
)

# Compute rates
vaccine_summary["deployment_rate_pct"] = (vaccine_summary["deployed"] / vaccine_summary["allocated"].replace(0, pd.NA)) * 100
vaccine_summary["administration_rate_pct"] = (vaccine_summary["administered"] / vaccine_summary["deployed"].replace(0, pd.NA)) * 100
vaccine_summary["uptake_rate_pct"] = (vaccine_summary["administered"] / vaccine_summary["allocated"].replace(0, pd.NA)) * 100
vaccine_summary["undeployed_stock"] = vaccine_summary["allocated"] - vaccine_summary["deployed"]
vaccine_summary["stock_in_country_not_admin"] = vaccine_summary["deployed"] - vaccine_summary["administered"]

vaccine_summary.head()
```

```python
plt.figure(figsize=(10,6))
sns.scatterplot(
    data=vaccine_summary,
    x="allocated",
    y="uptake_rate_pct",
    size="administered",      # bubble size = actual vaccines administered
    hue="Country",
    palette="tab10",
    sizes=(50, 500),
    legend="brief"
)
plt.title("Vaccine Allocation vs Uptake Rate by Country")
plt.xlabel("Allocated Doses")
plt.ylabel("Uptake Rate (%)")
plt.legend(bbox_to_anchor=(1.05, 1), loc="upper left")  # move legend outside
plt.tight_layout()
plt.show()
```

Uganda's vaccine uptake is lower than all other countries despite being a hotspot, whilst Sierra despite having low vaccine allocation has a high vaccine uptake


```python
# Prepare heatmap data
heatmap_data = vaccine_summary[["Country", "undeployed_stock", "stock_in_country_not_admin"]].set_index("Country")

plt.figure(figsize=(10,6))
sns.heatmap(
    heatmap_data,
    annot=True,
    fmt=".0f",
    cmap="Reds",
    linewidths=0.5
)
plt.title("Vaccine Stock Bottlenecks: Central vs In-Country")
plt.ylabel("Country")
plt.xlabel("Stock Type")
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(10,6))
sns.lineplot(data=vaccine_summary, x="Country", y="deployed", marker="o", label="Deployed")
sns.lineplot(data=vaccine_summary, x="Country", y="administered", marker="o", label="Administered")
plt.title("Vaccine Deployed vs Administered by Country")
plt.ylabel("Number of Doses")
plt.xticks(rotation=45)
plt.legend()
plt.tight_layout()
plt.show()
```

```python
# Compute allocation per 1,000 confirmed cases
mpox_cases = mpox.groupby("Country").agg(confirmed_cases=("Confirmed_Cases", "sum")).reset_index()
vaccine_summary = vaccine_summary.merge(mpox_cases, on="Country", how="left")
vaccine_summary["allocation_per_1000_cases"] = (vaccine_summary["allocated"] / vaccine_summary["confirmed_cases"]) * 1000

plt.figure(figsize=(10,6))
sns.scatterplot(
    data=vaccine_summary,
    x="confirmed_cases",
    y="allocation_per_1000_cases",
    size="allocated",
    hue="Country",
    palette="tab10",
    sizes=(50, 500),
    legend="brief"
)
plt.title("Vaccine Allocation Efficiency per 1,000 Cases")
plt.xlabel("Confirmed Cases")
plt.ylabel("Allocated Doses per 1,000 Cases")
plt.legend(bbox_to_anchor=(1.05, 1), loc="upper left")
plt.tight_layout()
plt.show()
```

Vaccine allocation efficiency is terribly low for Uganda, Burundi and Sierra Leone and could be the reason why they are a hotspot

```python
Objective 3:--Evaluate Surveillance and Workforce Capacity
Analyze the presence of active surveillance sites, testing laboratories, and trained vs deployed CHWs.
Identify whether stronger surveillance correlates with lower CFR or earlier detection.
Point out countries under-equipped relative to their case burden.
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

#DataFrame is called mpox
df = mpox.copy()

#Aggregate country-level surveillance and workforce metrics
country_summary = (
    df.groupby("Country")
    .agg(
        total_cases=("Confirmed_Cases", "sum"),           # fixed column name
        total_deaths=("Deaths", "sum"),
        total_active_sites=("Active_Surveillance_Sites", "sum"),  # fixed column name
        total_testing_labs=("Testing_Laboratories", "sum"),        # fixed column name
        total_trained_chws=("Trained_CHWs", "sum"),
        total_deployed_chws=("Deployed_CHWs", "sum")
    )
    .reset_index()
)

#Calculate CFR correctly
country_summary["cfr_percent"] = (country_summary["total_deaths"] / 
                                 country_summary["total_cases"].replace(0, pd.NA)) * 100  

# Compute per-case ratios for surveillance and workforce
country_summary["deployed_chws_per_case"] = country_summary["total_deployed_chws"] / country_summary["total_cases"]
country_summary["trained_chws_per_case"] = country_summary["total_trained_chws"] / country_summary["total_cases"]
country_summary["surveillance_per_case"] = (country_summary["total_active_sites"] + country_summary["total_testing_labs"]) / country_summary["total_cases"]

country_summary.head()  # Check the results
```

```python
#Lollipop chart: deployed CHWs per case
country_summary_sorted = country_summary.sort_values("deployed_chws_per_case")

plt.figure(figsize=(12,6))
plt.stem(
    country_summary_sorted["Country"],
    country_summary_sorted["deployed_chws_per_case"],
    linefmt='dodgerblue', markerfmt='o', basefmt=" "
)
plt.xticks(rotation=45)
plt.ylabel("Deployed CHWs per Confirmed Case")
plt.title("Country-level Workforce Coverage (CHWs per Case)")
plt.tight_layout()
plt.show()
```

```python
# cfr
country_summary["avg_cfr"] = country_summary["total_deaths"] / country_summary["total_cases"] * 100

# 4️⃣ Scatter plot: Surveillance vs CFR
plt.figure(figsize=(10,6))
sns.scatterplot(
    data=country_summary,
    x="surveillance_per_case",
    y="avg_cfr",           # now exists
    size="total_cases",    # bubble size = total cases
    hue="Country",
    sizes=(50, 500),
    alpha=0.7,
    palette="tab20"
)
plt.xlabel("Surveillance Sites + Labs per Case")
plt.ylabel("Average Case Fatality Rate (%)")
plt.title("Surveillance Capacity vs CFR (Bubble size = Total Cases)")
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left')  # moves legend outside
plt.tight_layout()
plt.show()
```

```python
#Heatmap: Workforce per-case ratios
workforce_metrics = country_summary.set_index("Country")[["deployed_chws_per_case", "trained_chws_per_case", "surveillance_per_case"]]
plt.figure(figsize=(12,6))
sns.heatmap(workforce_metrics, annot=True, fmt=".2f", cmap="YlOrRd")
plt.title("Per-case Workforce & Surveillance Metrics by Country")
plt.show()
```

```python
# --- Compute Priority Index (higher = more urgent) ---
country_metrics["priority_index"] = (
    (country_metrics["total_cases"] * 0.4) +
    (country_metrics["total_deaths"] * 0.3) +
    ((100 - country_metrics["latest_coverage"]) * 0.1) +
    ((1 / country_metrics["deployed_chws_per_case"].replace(0, pd.NA)) * 0.1) +
    ((1 / country_metrics["labs_per_case"].replace(0, pd.NA)) * 0.1)
)
```

```python
# ---Aggregate country-level metrics ---
country_metrics = df.groupby("Country").agg(
    total_cases=("Confirmed_Cases", "sum"),
    total_deaths=("Deaths", "sum"),
    latest_coverage=("Vaccine_Coverage", "max"),
    total_trained_chws=("Trained_CHWs", "sum"),
    total_deployed_chws=("Deployed_CHWs", "sum"),
    total_labs=("Testing_Laboratories", "sum")
).reset_index()

# Compute per-case ratios
country_metrics["deployed_chws_per_case"] = country_metrics["total_deployed_chws"] / country_metrics["total_cases"]
country_metrics["trained_chws_per_case"] = country_metrics["total_trained_chws"] / country_metrics["total_cases"]
country_metrics["labs_per_case"] = country_metrics["total_labs"] / country_metrics["total_cases"]
country_metrics["cfr_percent"] = (country_metrics["total_deaths"] / country_metrics["total_cases"]) * 100
```

```python
# ---Lollipop chart: priority index ---
country_priority_sorted = country_metrics_sorted.head(10)  # top 10 priority
plt.figure(figsize=(12,6))
plt.stem(
    country_priority_sorted["Country"],
    country_priority_sorted["priority_index"],
    linefmt='darkorange',
    markerfmt='o',
    basefmt=" "
)
plt.xticks(rotation=45)
plt.ylabel("Priority Index")
plt.title("Top 10 Priority Countries (Lollipop Chart)")
plt.tight_layout()
plt.show()
```

Some countries are already burning (hotspots), others have sparks (high priority index). Workforce and vaccines are like firebreaks,we must reinforce them to prevent new hotspots. Hotspots need immediate intervention (Burundi, Uganda, Sierra Leone). Countries like South Africa and Mozambique are watch closely: high priority index means they could escalate without sustained response. Workforce and vaccination are mitigating factors,strengthening them is an effective prevention strategy

```python
import pandas as pd
import numpy as np

df = mpox.copy()  # your original dataframe

# Normalize column names
df.columns = (
    df.columns.str.strip()
              .str.lower()
              .str.replace(r'[\s]+', '_', regex=True)
)

# Fix common misspellings / variants into a single canonical schema
rename_map = {
    'confirmed_cases': 'confirmed_cases',
    'deaths': 'deaths',
    'vaccinations_administered': 'vaccinations_administered',
    'active_surveillance_sites': 'active_surveillance_sites',
    'suspected_cases': 'suspected_cases',
    'case_fatality_rate': 'case_fatality_rate',
    'clade': 'clade',
    'weekly_new_cases': 'weekly_new_cases',
    'vaccine_dose_allocated': 'vaccine_dose_allocated',
    'vaccine_dose_deployed': 'vaccine_dose_deployed',
    'vaccine_coverage': 'vaccine_coverage',
    'testing_laboratries': 'testing_laboratories',   # <- typo fix
    'testing_laboratories': 'testing_laboratories',
    'trained_chws': 'trained_chws',
    'deployed_chws': 'deployed_chws',
    'surveillance_notes': 'surveillance_notes',
    'country': 'country',
    'report_date': 'report_date'
}
df = df.rename(columns=rename_map)

# Ensure dtypes are right
df['report_date'] = pd.to_datetime(df['report_date'], errors='coerce')
for c in ['confirmed_cases','deaths','vaccinations_administered',
          'active_surveillance_sites','suspected_cases','weekly_new_cases',
          'vaccine_dose_allocated','vaccine_dose_deployed','vaccine_coverage',
          'testing_laboratories','trained_chws','deployed_chws']:
    if c in df.columns:
        df[c] = pd.to_numeric(df[c], errors='coerce')
```
